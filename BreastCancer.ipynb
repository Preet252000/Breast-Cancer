```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_breast_cancer

C_SIZE_OF = 200                                             #function for number of samples we want
```

```python
X, y = load_breast_cancer(return_X_y= True, as_frame= True) #X is the data on two axis where y is the label to X
                                                            #return_X_y makes it data feature and as_frame makes it label, since we need it seperate

X['class'] = y                                              # y is defined as a lable to dataframe X as a binary function  0=benign 1=malignant

data = X[['mean radius','mean texture', 'class']]           #this functions defines to make a table with these 3 variables

dataset_sample = data.sample(C_SIZE_OF, random_state= 42)   #dataset_sample=name of the table, data.sample used to create dataframe, C_SIZE_OF= variable for number of smaples to be selected

```

```python
dataset_sample.head()                                       # '.' after dataset_dample is used to generate rows from dataset
```

```python
sns.scatterplot(data= dataset_sample, x='mean radius', y='mean texture', hue='class')
```

```python
sns.set_style('whitegrid')                                      #this function is used to make whitegrid
sns.scatterplot(data= dataset_sample, x='mean radius', y='mean texture', hue='class')

```

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_breast_cancer
C_SIZE_OF = 20
```

```python
X, y = load_breast_cancer(return_X_y= 'True', as_frame= 'true')
X['class']= y
data = X[['mean radius', 'mean texture', 'class']]

dataset_sample = data.sample(C_SIZE_OF, random_state=20)
```

```python
data= dataset_sample.head()
```

```python
sns.set_theme()
sns.scatterplot(data=dataset_sample, x='mean radius', y='mean texture', hue='class')
```

```python
sns.set_style('whitegrid')
sns.pairplot(data= dataset_sample, hue='class', vars=['mean radius', 'mean texture'])
```
